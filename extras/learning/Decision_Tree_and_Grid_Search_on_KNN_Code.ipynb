```python
import numpy
from sklearn import datasets, metrics
from sklearn.tree import DecisionTreeClassifier

X,y=datasets.load_iris(return_X_y=True)
print(X)
print(y)

X_train=X[range(0,150,2),:]
y_train=y[range(0,150,2)]

X_test=X[range(1,150,2),:]
y_test=y[range(1,150,2)]

clf = DecisionTreeClassifier(criterion='entropy') #creating instance of a classifier
#default criterion is gini

clf.fit(X_train, y_train)
prediction = clf.predict(X_test) #predict using the learnt classifier

print("############### Predictions #################")
print(prediction)
print("#############################################")

print("Accuracy:",metrics.accuracy_score(y_test, prediction, normalize=True))

print(metrics.classification_report(y_test, prediction))
print(metrics.confusion_matrix(y_test, prediction))
```

```python
import numpy
from sklearn import neighbors, datasets, metrics
from sklearn.model_selection import GridSearchCV

X, y=datasets.load_iris(return_X_y=True) #Loading the Iris Dataset

X_train=X[range(0,150,2),:]
y_train=y[range(0,150,2)]

X_test=X[range(1,150,2),:]
y_test=y[range(1,150,2)]

#creating instance of a classifier
clf = neighbors.KNeighborsClassifier()

#param_grid should be dictionary or list of dictionaries
#Specify hyperparameters which you want to tune

#If some hyperparameter you don't want to tune, specify that parameter
#while creating instance of the classifier

param_grid = [
  {'n_neighbors': [1, 5, 10, 20], 'weights': ['uniform', 'distance']} ]

clf_gs=GridSearchCV(clf,param_grid,cv=5)
#there is a refit argument whose default is True

clf_gs.fit(X_train, y_train)

#Print best parameters
print(clf_gs.best_params_)

prediction = clf_gs.predict(X_test) #predict using the learnt classifier

print("############### Predictions #################")
print(prediction)
print("#############################################")

print("Accuracy = ", metrics.accuracy_score(y_test, prediction, normalize=True))

print(metrics.classification_report(y_test, prediction))

print(metrics.confusion_matrix(y_test, prediction))
```
